```python
#Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Criando nosso Dataframe
df = pd.read_excel('T:/Inteligencia/2022/LUCIO/Github/Desafio_1/2be7520c-94cd-463f-8432-d580d0810344/Cusro_Python_Pandas_Digital_Innovation-master/datasets/AdventureWorks.xlsx')
```

```python
#Visualizando as 5 primeiras linhas
df.head()
```

```python
#Quantidade de linhas e colunas
df.shape
```

```python
#Verificando os tipos de dados
df.dtypes
```

```python
#Qual a Receita Total?
df['Valor Venda'].sum()
```

```python
#Qual o Custo Total?
df['custo'] = df['Custo Unitário'].mul(df['Quantidade']) #Criando a coluna custo
df.head(1)
```

```python
#Qual o Custo Total?
round(df['custo'].sum(), 2)#Arrodondar em duas casas decimais
```

```python
#Qual é o Lucro?
df['lucro'] = df['Valor Venda'] - df['custo']
df.head(1)
```

```python
#Qual é o Lucro Total?
round(df['lucro'].sum(), 2)
```

```python
#Criando uma coluna com a quantidade de dias para entregar o produto
df['Tempo_envio'] = df['Data Envio'] - df['Data Venda']
df.head(1)
```

```python
#Calcular a Média do Tempo de Envo para cada marca
#Para isso devemos transformar a coluna Tempo_envio em númerica
df['Tempo_envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
df.head(1)
```

```python
#Conferir tipo dos dados
df.dtypes
```

```python
#Média do tepo de envios por marca
df.groupby('Marca')['Tempo_envio'].mean()
```

```python
#Verificar valores missing
df.isnull().sum()
```

```python
#Saber lucro por Ano e por Marca
df.groupby([df['Data Venda'].dt.year, "Marca"])['lucro'].sum()
```

```python
#Retirar a visualização dos dados em noação científica
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Saber lucro por Ano e por Marca
df.groupby([df['Data Venda'].dt.year, "Marca"])['lucro'].sum()
```

```python
#Resetando index e apresentar agrupamento como dataframe
lucro_ano = df.groupby([df['Data Venda'].dt.year, "Marca"])['lucro'].sum().reset_index()
lucro_ano
```

```python
#Qual o total de produtos vendidos?
df.groupby("Produto")['Quantidade'].sum().sort_values(ascending=False)
```

```python
#Gráfico de barras horizontais do Total de Produtos Vendidos
df.groupby("Produto")['Quantidade'].sum().sort_values(ascending=True).plot.barh(title="Total Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
#Gráfico de barras horizontais do Lucro por Produtos
df.groupby("Produto")['lucro'].sum().sort_values(ascending=True).plot.barh(title="Lucro x Produtos")
plt.xlabel("Lucro")
plt.ylabel("Produto");
```

```python
#Gráfico do Lucro por Ano
df.groupby(df['Data Venda'].dt.year)['lucro'].sum().plot.bar(title="Lucro x Ano")
plt.xlabel('Ano')
plt.ylabel('Receita');
```

```python
#Lucro por Ano descritivo
df.groupby(df['Data Venda'].dt.year)['lucro'].sum()
```

```python
#Selecionando apenas as vendas de 2009
df_2009 = df[df['Data Venda'].dt.year == 2009]
df_2009.head()
```

```python
#Gráfico de linhas do lucro por mês
df_2009.groupby(df_2009['Data Venda'].dt.month)['lucro'].sum().plot(title='Lucro x Mês')
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

```python
#Gráfico de barras do lucro por marca
df_2009.groupby("Marca")["lucro"].sum().plot.bar(title='Lucro por Marca')
plt.xlabel('Marca')
plt.ylabel('Lucro')
plt.xticks(rotation = 'horizontal'); #rotaciona as legendas das barras na posição horizontal
```

```python
#Gráfico de barras do lucro por classe

df_2009.groupby('Classe')['lucro'].sum().plot.bar(title="Lucro x Classe")
plt.xlabel("Classe")
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```
